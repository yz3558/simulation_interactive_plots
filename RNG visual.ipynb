```python
#CODING NOT REQUIRED
import numpy as np
from ipywidgets import widgets
import plotly.graph_objs as go
 
    



def generate(N,a0,m0,x0,c0):
    a=a0
    m=m0
    x=x0
    c=c0
    rd = np.repeat(0.0,N)
    for i in range(N):
      x = (x*a+c)%m
      rd[i]=x/m
    return rd


def initialize():
    n = widgets.IntSlider(
        value=1.0,
        min=1.0,
        max=50.0,
        step=1.0,
        description='# of samples:',
        continuous_update=False
    )
    c=widgets.IntText(
        value=0,
        description='c value:',
        disabled=False
    )

    
    a=widgets.IntText(
        value=7,
        description='a value:',
        disabled=False
    )
    
    

    m=widgets.IntText(
        value=40,
        description='m value:',
        disabled=False
    )

    x0=widgets.IntText(
        value=3,
        description='x0 (seed):',
        disabled=False
    )

    settingx=widgets.Dropdown(
        options=["customize",'set1', 'set2'],
        value='customize',
        description='fix algorithm:',
        disabled=False,
    ) 
    rd=generate(n.value,a.value,m.value,x0.value,c.value)
    y=np.zeros(len(rd))

    # Assign an empty figure widget with two traces
    trace1=go.Scatter(x=rd,y=y,mode='markers', hoverinfo="x", marker=dict(size=20))
    g = go.FigureWidget(data=[trace1],
                        layout=go.Layout(
                            title=dict(
                            text=''
                            )
                            ,xaxis = dict(title = "Values of random numbers", range = [-0.2,1.2]),
                            yaxis = dict(title = "", range = [-0.1,0.1],showticklabels=False),
                            hovermode=None,
                            margin={'l': 0, 'r': 0, 't': 20, 'b': 0},width=600, height=200
                            

                        ))

    return n,a,m,x0,settingx,g,c
    
n,a,m,x0,settingx,g,c=initialize()
 
def response(change ):
    if settingx.value=="set1":
        a.value=7
        m.value=40
        c.value=0
    elif settingx.value=="set2":
        a.value=7**5
        m.value=2**31-1
        c.value=0
    rd=generate(n.value,a.value,m.value,x0.value,c.value)
    y=np.zeros(len(rd))
    with g.batch_update():
        g.data[0].x = rd
        g.data[0].y = y

n.observe(response, names="value")
a.observe(response, names="value")
c.observe(response, names="value")
m.observe(response, names="value")
x0.observe(response, names="value")
settingx.observe(response,names="value")


container1 = widgets.HBox([n,settingx])
container2 = widgets.HBox([a,m])
container3 = widgets.HBox([c,x0])
Widget1=widgets.VBox([container1,container2,container3, g])
display(Widget1)

```
