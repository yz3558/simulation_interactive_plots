```python
import numpy as np
import plotly.graph_objs as go 
from ipywidgets import widgets

 
import scipy.stats as spst
 

next_arrival=widgets.Button(
    description="next arrival")
 

clear=widgets.Button(
    description="clear")

 

x=np.array([0,5])
y=np.array([0,0])
at_x=np.array([0])
at_y=np.array([0])
at_string=[""]

trace1=go.Scatter(x=x,y=y, 
              hoverinfo="text",text="",name="time between 0 and 5",mode="lines",marker=dict(size=10,color="yellow") )

trace2=go.Scatter(x=at_x[1:-1],y=at_x[1:-1], 
              hoverinfo="text",text=at_string[1:-1],name="previous arrivals",mode="markers" ,marker=dict(size=10,color="blue") )

trace3=go.Scatter(x=[at_x[-1]],y=[at_x[-1]],hoverinfo="text", text=[at_string[-1]],
                  marker=dict(size=10,color="red"),name="new arrival",mode="markers" )

g = go.FigureWidget(data=[trace1],
                        layout=go.Layout(
                        title=dict(
                        text='homogeneous Poisson arrival'
                        ),
                
                        xaxis = dict(title = " time space", range = [-0.1,8],showticklabels=True),
                        yaxis = dict(title = "", range = [-.1,0.5] ,showticklabels=False),
                        hovermode=None,
                        margin={'l': 0, 'r': 0, 't': 0, 'b': 0},width=1000, height=200  ),
                   )
g.update_layout(
                     title_x=0.5,
                 title_y=0.9,

                 legend=dict(

                    x=0.7,
                    y=0.7,
                    traceorder="normal",
                    font=dict(
                        family="sans-serif",
                        size=12,
                        color="black"
                    ))
                   
)

lm=1
def response2(change):
    global at_x
    global at_y
    at_x=np.append(at_x,at_x[-1]-1/lm*np.log(np.random.rand()))
    at_y=np.append(at_y,0)
    at_string=np.round(at_x,2).tolist()
    at_string=["arrival time:"+str(item) for item in at_string]
    if  at_x[-1]<5:
        with g.batch_update():
            g.data = [g.data[0]]
            trace2.text=at_string[1:-1]
            trace2.x=at_x[1:-1]
            trace2.y=at_y[1:-1]
            trace3.text=[at_string[-1]]
            trace3.x=[at_x[-1]]
            trace3.y=[at_y[-1]]
            g.add_trace(trace2)
            g.add_trace(trace3)
    else:
        next_arrival.disabled=True
        with g.batch_update():
            g.data = [g.data[0]]
            trace2.text=at_string[1:-1]
            trace2.x=at_x[1:-1]
            trace2.y=at_y[1:-1]
            trace3.text=[at_string[-1]]
            trace3.x=[at_x[-1]]
            trace3.y=[at_y[-1]]
            trace3.marker.symbol="x-open"
            trace3.marker.color="black"
            trace3.name="rejected arrival"

            g.add_trace(trace2)
            g.add_trace(trace3)
        
def response1(change):
    global at_x
    at_x=np.array([0])
    global at_y
    at_y=np.array([0])
    next_arrival.disabled=False
    trace3.marker.symbol="circle"
    trace3.marker.color="red"
    trace3.name="new arrival"

    g.data = [g.data[0]]
 

        
    
container1 = widgets.HBox([next_arrival,clear]) 

container2 = widgets.HBox([g ])

 
next_arrival.on_click(response2)
clear.on_click(response1)

Widget1=widgets.VBox([container1,container2] )
Widget1
```
